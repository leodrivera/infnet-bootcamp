# LAB 2 - DATA WRANGLING

```
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

# Read the online file by the URL provides above, and assign it to variable "df"
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df_wrangling = pd.read_csv(url, names=headers)
```

```
df_wrangling.replace("?", np.nan, inplace = True)
df_wrangling.head(5)
```

```
missing_data = df_wrangling.isnull()
missing_data.head(5)
```

```
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")  
```

<h4>Calculate the average of the column </h4>


```
avg_norm_loss = df_wrangling["normalized-losses"].astype("float").mean()
avg_norm_loss
```

<h4>Replace "NaN" by mean value in "normalized-losses" column</h4>


```
df_wrangling.head()
```

```
df_wrangling["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```
df_wrangling.head()
```

```
avg_stroke = round(df_wrangling["stroke"].astype("float").mean(),2)
avg_stroke
```

```
df_wrangling["stroke"].replace(np.nan, avg_stroke, inplace=True)
```

```
avg_bore = round(df_wrangling["bore"].astype("float").mean(),2)
avg_bore
```

```
df_wrangling["bore"].replace(np.nan, avg_bore, inplace=True)
```

```
avg_horsepower = round(df_wrangling["horsepower"].astype("float").mean())
avg_horsepower
```

```
df_wrangling["horsepower"].replace(np.nan, avg_horsepower, inplace=True)
```

```
avg_peakrpm = round(df_wrangling["peak-rpm"].astype("float").mean())
avg_peakrpm
```

```
df_wrangling["peak-rpm"].replace(np.nan, avg_peakrpm, inplace=True)
```

```
freq_num_of_doors = df_wrangling['num-of-doors'].value_counts().idxmax()
freq_num_of_doors
```

```
df_wrangling["num-of-doors"].replace(np.nan, freq_num_of_doors, inplace=True)
```

```
df_wrangling.dropna(subset=["price"], inplace=True)
```

```
df_wrangling.isnull().sum()
```

```
df_wrangling.reset_index(drop=True, inplace=True)
```

## Data Standardization

```
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df_wrangling['city-L/100km'] = 235/df_wrangling["city-mpg"]

# check your transformed data 
df_wrangling.head()
```

```
# transform mpg to L/100km by mathematical operation (235 divided by mpg)
df_wrangling["highway-mpg"] = 235/df_wrangling["highway-mpg"]

# rename column name from "highway-mpg" to "highway-L/100km"
df_wrangling.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)

# check your transformed data 
df_wrangling.head()
```
